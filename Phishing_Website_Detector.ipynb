# Cleaning the Class Labels and Inspecting for Missing Valus

It is not a good practice to build machine learning models where the labels are encoded as negative values as it affects the performance of the models. Therefore, we will need to change all the -1 values in the "Result" column to 0.

```python
data['Result'].replace(-1, 0, inplace=True)
```

Next, we should check what values does the "Result" column of the dataset contain.

```python
data['Result'].unique()
```

It is clear that all the -1 values in the "Result" column was replaced with 0 as the "Result" column now only contains 0 and 1 as possible values.

Next, we should check whether any of the feature suffers from the problem of missing values. First, we should check all the different values that each column contains.

```python
# checking if the dataset has missing values
data.isna().sum()
```

Although the `isnull` method effectively does the same thing as the `isna` method, we can look for missing values in a different way as below.

```python
data.isnull().any()
```

Hence, there are no missing values in the for null values or NaN values. 

We should also look for missing values in different forms other than null values and NaN values. This can be done by checking all the different values of each feature in the dataset.

```python
for feature in data.columns:
    print(f"{feature}: {data[feature].unique()}")
```

Since all of the features have the expected number of different values, it is clear that there are no missing values in the dataset.

Next, we need to split the dataset into train and test sets. This can be done using methods from the scikit-learn library. The split should be 80:20. Also, since we want reproducible outputs across multiple function calls, we should pass an int to the parameter `random_state`.

```python
# import the scikit-learn library
from sklearn.model_selection import train_test_split

x = data.iloc[:,0:30].values.astype(int)  # predictors
y = data.iloc[:,30].values.astype(int)  # class labels
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=np.random.seed(8))
```

# Training a model and evaluating performance
As we saw earlier, the class labels have only two possible values 0 and 1. Therefore, it is clear that we are trying to solve a classification problem so we can train a logistic regression model.

```python
# filter the unnecessary warnings
import warnings
warnings.filterwarnings("ignore")
```

First, we need to import all the necessary libraries, functions, etc.

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
from sklearn.linear_model import LogisticRegression
!pip install wandb
import wandb
import time
```

Next, we can define a utility function for training the logistic regression model while also measuring its training time and performance.

```python
def train_eval_pipeline(model, train_data, test_data, name):
    # initialize weights and biases
    wandb.init(project="phishing-website-detection", name=name)
    
    # segregate the datasets
    (x_train, y_train) = train_data
    (x_test, y_test) = test_data
    
    # train the model and log all the necessary metrics
    start = time.time()
    model.fit(x_train, y_train)
    end = time.time() - start
    prediction = model.predict(x_test)

    wandb.log({"accuracy":accuracy_score(y_test, prediction)*100.0,\
               "precision": precision_recall_fscore_support(y_test, prediction, average='macro')[0],
               "recall": precision_recall_fscore_support(y_test, prediction, average='macro')[1],
               "training_time":end})
    
    print("\nAccuracy score of the Logistic Regression classifier with default hyperparameter values {0:.2f}%"\
              .format(accuracy_score(y_test, prediction)*100.))
    print("\n")
    print("---Classification report of the Logistic Regression classifier with default hyperparameter value---")
    print("\n")
    print(classification_report(y_test, prediction, target_names=["Phishing Websites", "Normal Websites"]))
```

Next, we need to train a logistics regression and evaluate its performance by using the `train_eval_pipeline` function.

```python
log_reg_model = LogisticRegression()
train_eval_pipeline(log_reg_model, (x_train, y_train), (x_test, y_test), "logistic_regression")
```

From the above classification report, we can see that the model's precision, recall, and f1-score is high (greater than 0.90) for both phishing websites and normal websites. This tells us that the current model makes fairly good predicitons. 

Next, in order to improve the logistic regression model, we can tune the hyperparameters of the model so that they are searched in a randomized fashion. The grid of hyperparameters can be constructed as below.

```python
# define the grid of values
penalty = ['l1', 'l2']
C = [0.8, 0.9, 1.0]
tol = [0.01, 0.001, 0.0001]
max_iter = [100, 150, 200, 250]

parameter_grid = dict(penalty=penalty, C=C, tol=tol, max_iter=max_iter)
```

Next, we can perform a random search process. We need to fit the `RandomizedSearchCV` model to the training data and find the best estimator and its hyperparameter values. 

```python
from sklearn.model_selection import RandomizedSearchCV


# instantiate RandomizedSearchCV with the required parameters
random_model = RandomizedSearchCV(estimator=log_reg_model, param_distributions=parameter_grid, cv=5)

# fir random_model to the training data
random_model_result = random_model.fit(x_train, y_train)

# summarize results
best_score, best_params = random_model_result.best_score_, random_model_result.best_params_
print(f"Best score: {best_score*100:.2f} using {best_params}")
```

As we can see, the random search did not improve the performance of the model greatly. 

We can further check if random search helps in boosting up the accuracy score by taking the hyperparameter values we got from the random search process and train another logistic regression model with those values.

First, we can log the hyperparameter values with which we are going to train the model.

```python
config = wandb.config
config.tol = 0.0001
config.penalty = "l2"
config.C = 0.8
```

Now we can train the logistic regression one last time using the hyperparameter values we found.

```python
# train the model
log_reg_final = LogisticRegression(tol=config.tol, penalty=config.penalty, max_iter=250, C=config.C)
train_eval_pipeline(log_reg_final, (x_train, y_train), (x_test, y_test), "logistic-regression-random-search")
```

To conclude, random search of hyperparameter values did not help to improve the accuracy score.
